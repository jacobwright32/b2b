```python
import os
from os import listdir
from os.path import isfile, join
from fpdf import FPDF
mypath = 'Cell'
from PIL import Image
```

```python
cell_path = []
for i in [x[0] for x in os.walk('Cell')]:
    if 'check' in i:
        continue
    else:
        cell_path.append(i)
```

```python
for mypath in cell_path:
    print(mypath)
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    if len(mypath.split('/')) > 1:
        cell = mypath.split('/')[1]
    else:
        cell = 'Cell'
    pdf = FPDF()
    idx = 0
    for image in onlyfiles:
        idx+=1
        print(image,idx/len(onlyfiles)*100,'%')
        if cell == 'Cell':
            cover = Image.open(f'Cell/{image}')
        else:
            cover = Image.open(f'Cell/{cell}/{image}')
        width, height = cover.size
        pdf.add_page(orientation='L',format=(height,width))
        if cell == 'Cell':
            pdf.image(f'Cell/{image}',0,0,float(width), float(height))
        else:
            pdf.image(f'Cell/{cell}/{image}',0,0,float(width), float(height))
    if cell != 'Cell':
        pdf.output(f"{cell}.pdf", "F")
    else:
        pdf.output(f"{cell}_agg_view.pdf", "F")
    print('Done===========================================================================')
```
